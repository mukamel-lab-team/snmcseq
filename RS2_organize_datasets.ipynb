```python
# check if all datasets are good dataset
with snmcseq_utils.cd(PATH_DATASETS):
    datasets = sorted(glob.glob('*'))
for dataset in datasets:
    with snmcseq_utils.cd(os.path.join(PATH_DATASETS, dataset)):
        try:
            df_meta = pd.read_table('mapping_summary_{}.tsv'.format(dataset))
            n_cells = df_meta.shape[0]
        except:
            print(dataset, "Error in opening mapping summary")
            continue
        
        n_allc = len(glob.glob('./allc/allc_*.tsv.bgz'))
        n_allc_tbi = len(glob.glob('./allc/allc_*.tsv.bgz.tbi'))
        
        n_binc = len(glob.glob('./binc/binc_*.tsv.bgz'))
        n_genebody = len(glob.glob('./gene_level/genebody_*.tsv.bgz'))
        
        flag = 0
        if n_cells != n_allc:
            print(dataset, n_cells, 'Error allc! ({})'.format(n_allc))
            flag += 1
        if n_cells != n_allc_tbi:
            print(dataset, n_cells, 'Error allc tabix! ({})'.format(n_allc_tbi))
            flag += 1
        if n_cells != n_binc:
            print(dataset, n_cells, 'Error binc! ({})'.format(n_binc))
            flag += 1
        if n_cells != n_genebody:
            print(dataset, n_cells, 'Error genebody! ({})'.format(n_genebody))
            flag += 1
        
        if flag == 0:
            pass
            print(dataset, n_cells, "All good!")
    
```
