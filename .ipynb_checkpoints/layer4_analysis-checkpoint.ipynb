```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import gaussian_kde

from sklearn.decomposition import PCA
from sklearn.decomposition import FastICA

from snmcseq_utils import plot_tsne_values
from snmcseq_utils import plot_tsne_labels
from snmcseq_utils import tsne_and_boxplot
```

```python
mc_c_file = './data/gene_level/genebody_mCH_human_combined_summary.tsv'
df_mc_c = pd.read_table(mc_c_file, index_col='id')
print(df_mc_c.shape)
```

```python
meta_file = './data/metadata/metadata_human_combined_updated.tsv' 
cluster_file = './data/cluster/cluster_MB_v1_MB_EA_MB_EB/clusters_binc_mCHmCG_human_combined_100000_summary_normalized.tsv'
tsne_file = './data/tsne/old/tsne_perp50_binc_mCHmCG_human_combined_100000_summary_normalized.tsv'
gene_id_name_file = './data/references/gene_id_to_names.tsv'

df_meta = pd.read_table(meta_file, index_col='Sample')
df_cluster = pd.read_table(cluster_file, index_col='sample')
df_tsne = pd.read_table(tsne_file, index_col='sample')
df_gene = pd.read_table(gene_id_name_file, index_col='geneID')

print(df_meta.shape)
print(df_tsne.shape)
print(df_cluster.shape)
print(df_gene.shape)
```

```python
df_info = pd.merge(df_meta, df_cluster, left_index=True, right_index=True)
df_info = pd.merge(df_info, df_tsne, left_index=True, right_index=True)
print(df_info.shape)


cells = df_info[df_info.cluster_ID=='cluster_52'].index.values
cells_1 = df_info[(df_info.cluster_ID=='cluster_52') & (df_info.Biosample=='MB_v1')].index.values
cells_2 = df_info[(df_info.cluster_ID=='cluster_52') & (df_info.Biosample=='MB_EA')].index.values
cells_3 = df_info[(df_info.cluster_ID=='cluster_52') & (df_info.Biosample=='MB_EB')].index.values
print(cells.shape)
print(cells_1.shape)
print(cells_2.shape)
print(cells_3.shape)

df_mc = df_mc_c.filter(regex='_mc')[cells + '_mc']
df_c = df_mc_c.filter(regex='_c')[cells + '_c']

# get df_mcc
df_mc.columns = [col[:-len('_mc')] for col in df_mc.columns]
df_c.columns = [col[:-len('_c')] for col in df_c.columns]
# df_mcc = df_mc/df_c
# df_mcc.columns = [col+'_mcc' for col in df_mcc.columns]
# print(df_mcc.shape)
```

```python
# get mc and c for all samples

df_sum = pd.DataFrame()
df_sum['MB_v1_mc'] = df_mc[cells_1].sum(axis=1)
df_sum['MB_EA_mc'] = df_mc[cells_2].sum(axis=1)
df_sum['MB_EB_mc'] = df_mc[cells_3].sum(axis=1)
df_sum['MB_v1_c'] = df_c[cells_1].sum(axis=1)
df_sum['MB_EA_c'] = df_c[cells_2].sum(axis=1)
df_sum['MB_EB_c'] = df_c[cells_3].sum(axis=1)

print(df_sum.shape)

# get df_mcc
df_sum_mc = df_sum.filter(regex='_mc$')
df_sum_c = df_sum.filter(regex='_c$')
df_sum_mc.columns = [col[:-len('_mc')] for col in df_sum_mc.columns]
df_sum_c.columns = [col[:-len('_c')] for col in df_sum_c.columns]
df_sum_mcc = df_sum_mc/df_sum_c
df_sum_mcc.columns = [col+'_mcc' for col in df_sum_mcc.columns]
print(df_sum_mcc.shape)
# sns.heatmap(df_sum_mcc)
# plt.show()
```

```python
# filter out genes with low coverage
cutoff_2 = 1000
cov = (df_sum_c > cutoff_2)*1
gene_cov = cov.apply(np.all, axis=1)
df_sum_mccf = df_sum_mcc.loc[gene_cov, :]
print(df_sum_mccf.shape)
# sns.heatmap(cov)
# plt.show()
```

```python
x = np.arange(0, 0.18, 0.01)
fig, axs = plt.subplots(1, 3, figsize=(12,6))
ax = axs[0]
ax.plot(x, x, '--', linewidth=1, color='C1')
xy = np.vstack([df_sum_mccf.MB_v1_mcc, df_sum_mccf.MB_EA_mcc]) 
z = gaussian_kde(xy)(xy)
ax.scatter(df_sum_mccf.MB_v1_mcc, df_sum_mccf.MB_EA_mcc, c=np.log10(z), s=5)
ax.set_xlabel('MB_v1')
ax.set_ylabel('MB_EA')

ax = axs[1]
ax.plot(x, x, '--', linewidth=1, color='C1')
xy = np.vstack([df_sum_mccf.MB_v1_mcc, df_sum_mccf.MB_EB_mcc]) 
z = gaussian_kde(xy)(xy)
ax.scatter(df_sum_mccf.MB_v1_mcc, df_sum_mccf.MB_EB_mcc, c=np.log10(z), s=5)
ax.set_xlabel('MB_v1')
ax.set_ylabel('MB_EB')

ax = axs[2]
ax.plot(x, x, '--', linewidth=1, color='C1')
xy = np.vstack([df_sum_mccf.MB_EA_mcc, df_sum_mccf.MB_EB_mcc]) 
z = gaussian_kde(xy)(xy)
ax.scatter(df_sum_mccf.MB_EA_mcc, df_sum_mccf.MB_EB_mcc, c=np.log10(z), s=5)
ax.set_xlabel('MB_EA')
ax.set_ylabel('MB_EB')
fig.tight_layout()
plt.show()
```

```python
df_sum_diff = pd.DataFrame()
df_sum_diff['diff_EA_v1'] = (df_sum_mccf.MB_EA_mcc - df_sum_mccf.MB_v1_mcc)
df_sum_diff['diff_EB_v1'] = (df_sum_mccf.MB_EB_mcc - df_sum_mccf.MB_v1_mcc)
df_sum_diff['diff_EB_EA'] = (df_sum_mccf.MB_EB_mcc - df_sum_mccf.MB_EA_mcc)


fig, axs = plt.subplots(1, 3, figsize=(16, 4))
ax = axs[0]
ax.plot(df_sum_diff.diff_EA_v1.apply(np.abs).sort_values(ascending=False).values, '-o', markersize=1, label='|MB_EA - MB_v1|')
ax.plot(df_sum_diff.diff_EB_v1.apply(np.abs).sort_values(ascending=False).values, '-o', markersize=1, label='|MB_EB - MB_v1|')
ax.plot(df_sum_diff.diff_EB_EA.apply(np.abs).sort_values(ascending=False).values, '-o', markersize=1, label='|MB_EB - MB_EA|')
ax.set_xlabel('Genes')
ax.set_ylabel('Differences in mCH level')
ax.legend()
ax = axs[1]
ax.plot(df_sum_diff.diff_EA_v1.apply(np.abs).sort_values(ascending=False).values[:100], '-o', markersize=5, label='|MB_EA - MB_v1|')
ax.plot(df_sum_diff.diff_EB_v1.apply(np.abs).sort_values(ascending=False).values[:100], '-o', markersize=5, label='|MB_EB - MB_v1|')
ax.plot(df_sum_diff.diff_EB_EA.apply(np.abs).sort_values(ascending=False).values[:100], '-o', markersize=5, label='|MB_EB - MB_EA|')
ax.set_xlabel('Genes (Top 100 most different ones)')
ax.set_ylabel('Differences in mCH level')
ax.legend()
ax = axs[2]
ax.plot(df_sum_diff.diff_EA_v1.apply(np.abs).sort_values(ascending=False).values[:10], '-o', markersize=5)
ax.plot(df_sum_diff.diff_EB_v1.apply(np.abs).sort_values(ascending=False).values[:10], '-o', markersize=5)
ax.plot(df_sum_diff.diff_EB_EA.apply(np.abs).sort_values(ascending=False).values[:10], '-o', markersize=5)
fig.tight_layout()
plt.show()
```

```python
# filter out genes with low coverage
# doing imputation

# cutoff = 100
# cov = (df_c >= cutoff)*1
# gene_cov = cov.apply(np.all, axis=1)
# df_mccf = df_mcc.loc[gene_cov, :]
# print(df_mccf.shape)


# remove low coverage gene

base_call_cutoff = 100 
sufficient_coverage_fraction = 0.98 
condition = (df_c > base_call_cutoff).sum(axis=1) >= sufficient_coverage_fraction*(df_c.shape[1]) 


print("Matrix size before pruning... "+ str(df_mc_c.shape))
df_f = df_mc_c.loc[condition]
print("Matrix size after pruning... "+ str(df_f.shape))

# get mcc matrix with kept bins and nan values for low coverage sites
df_c_nan = df_c.copy()
df_c_nan[df_c < base_call_cutoff] = np.nan
df_mcc = df_mc.loc[condition]/df_c_nan.loc[condition]
print(df_mcc.shape)

# imputation (missing value -> mean value of all cells)
print('Imputing data...')
means = df_mcc.mean(axis=1)
fill_value = pd.DataFrame({col: means for col in df_mcc.columns})
df_mcc.fillna(fill_value, inplace=True)

# add "_mcc" suffix
df_mcc.columns = df_mcc.columns.values + '_mcc'

# here it was called as df_mccf in the following
df_mccf = df_mcc.copy()
```

```python
# run pca
n_components = 10
pca = PCA(n_components=n_components)
pca_res = pca.fit(df_mccf.T)
pcs = pca.fit_transform(df_mccf.T)
print(pca_res.components_.shape)
print(pcs.shape)

assert (cells+'_mcc').tolist() == df_mccf.T.index.tolist()
df_pcs = pd.DataFrame(pcs, columns=['PC'+str(i+1) for i in range(n_components)], index=cells)
```

```python
# plot PCA results

df_plot_pcs = pd.merge(df_pcs, df_info, left_index=True, right_index=True)
df_plot_pcs = df_plot_pcs[df_pcs.columns[:4].tolist() + ['Biosample']]
print(df_plot_pcs.shape)

g = sns.pairplot(df_plot_pcs, hue='Biosample', diag_kind='kde', markers='o', size=2.5)
# print(g.axes.shape)
# for i, j in zip(*np.triu_indices_from(g.axes, k=1)):
#     g.axes[i, j].set_visible(False)
plt.show()

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
ax = axs[0]
xvalues = [i+1 for i in range(len(pca_res.explained_variance_ratio_))]
ax.plot(xvalues, pca_res.explained_variance_ratio_, 'o-')
ax.set_title('Explained variance ratio')
ax.set_xticks(xvalues)
ax = axs[1]
ax.plot(xvalues, np.cumsum(pca_res.explained_variance_ratio_), 'o-')
ax.set_title('Cumulative explained variance ratio')
ax.set_xticks(xvalues)
fig.savefig('./results/pca.pdf')
plt.show()
```

```python
# look at PC2 (which seperates MB_EA from other biosamples)
n_components = 10
df_pcc = pd.DataFrame(pca_res.components_.T, columns=['PC'+str(i+1) for i in range(n_components)], index=df_mccf.index)

fig, axs = plt.subplots(1, 3, figsize=(12, 6))
pcc3 = df_pcc.PC1.apply(np.abs).sort_values(ascending=False)
ax = axs[0]
ax.plot(pcc3.values, 'o-')
ax = axs[1]
ax.plot(pcc3.values[:100], 'o-')
ax = axs[2]
ax.plot(pcc3.values[:10], 'o-')
fig.tight_layout()
plt.show()
```

```python
df_pc3_top_genes = pd.merge(pcc3.to_frame(), df_gene, left_index=True, right_index=True)
df_pc3_top_genes.head(10)
```

```python
# making tsne and boxplots
# snmcseq tsne and boxplot function
n_top = 10 

plot_tsne_labels(df_info, tc='Biosample', t_xlim=[10, 30], t_ylim=[-15, 10], s=5, 
                output='./results/gene_tsne_and_boxplots/l4_biosamples.pdf')

tsne_and_boxplot(df_info[df_info.cluster_ID=='cluster_52'], tc='mCH/CH', bx='Biosample', by='mCH/CH', 
                 t_xlim=[10, 30], t_ylim=[-15, 10],
                 figsize=(8, 10),
                 title='Global mCH',   
                 output='./results/gene_tsne_and_boxplots/l4_global_mCH.pdf')

# idx is gene_id
for idx, row in df_mcc.loc[df_pc3_top_genes.head(n_top).index,:].iterrows():
    row.index = [index[:-len('_mcc')] for index in row.index]
    df_plot = pd.merge(row.to_frame(), df_info, left_index=True, right_index=True)
    gene_name = df_gene.loc[idx, 'geneName']
    col_name = gene_name + ' mCH'
    df_plot.rename(columns={idx: col_name}, inplace=True)
    tsne_and_boxplot(df_plot, tc=col_name, bx='Biosample', by=col_name, 
                     t_xlim=[10, 30], t_ylim=[-15, 10],
                    title='Genebody mCH: ' + gene_name, 
                    output='./results/gene_tsne_and_boxplots/l4_%s.pdf' % gene_name)
    
```

```python
# fastICA

n_components = 10
print(df_mccf.T.shape)

# run pca
pca = PCA(n_components=n_components)
pca_res = pca.fit(df_mccf.T)
pcs = pca.fit_transform(df_mccf.T)
print(pca_res.components_.shape)
print(pcs.shape)

assert (cells+'_mcc').tolist() == df_mccf.T.index.tolist()
df_pcs = pd.DataFrame(pcs, columns=['PC'+str(i+1) for i in range(n_components)], index=cells)

# run ica
ica = FastICA(n_components=n_components)
ica_res = ica.fit(pcs)
ics = ica.fit_transform(pcs)

print(ics.shape)

assert (cells+'_mcc').tolist() == df_mccf.T.index.tolist()
df_ics = pd.DataFrame(ics, columns=['IC'+str(i+1) for i in range(n_components)], index=cells)
```

```python
df_plot = pd.merge(df_info, df_ics, left_index=True, right_index=True)

for col in df_ics.columns:
    plot_tsne_values(df_plot, tc=col, title=col, 
                     t_xlim=[10, 30], t_ylim=[-15, 10],
                    )
```

```python
# look at IC5 (which seperates different biosamples)

n_components = 10
df_icc = pd.DataFrame(ica_res.components_.T, columns=['IC'+str(i+1) for i in range(n_components)])

fig, ax = plt.subplots(figsize=(8, 6))
icc5 = df_icc.IC6
ax.plot([i+1 for i in range(n_components)], icc5.apply(np.abs).values, 'o-')
ax.set_xticks([i+1 for i in range(n_components)])
ax.set_xlabel('PCs')
plt.show()
```
