```python
# summarize mouse published
meta_path = os.path.join(PATH_REFERENCES, 'Mouse_published/summary_mouse_published.tsv')
genes_path = '/cndd/Public_Datasets/single_cell_methylome/gene_level/mouse/combined_sample_mCH_genebody.txt' 
bins_path = '/cndd/Public_Datasets/single_cell_methylome/binc/mouse_combined_samples_CH_100000.tsv'
```

```python
df_meta = pd.read_table(meta_path, index_col='Sample') 
print(df_meta.shape)
df_meta.head()
```

```python
bmcc = pd.read_table(bins_path, 
                    index_col=['chr', 'bin'], dtype={'chr': object, 'bin': int}
                    )
print(bmcc.shape)
bmcc.head()
```

```python
df_c = bmcc.filter(regex='_c$')
df_mc = bmcc.filter(regex='_mc$')
print(df_c.shape)
print(df_mc.shape)

```

```python
# new_cols = []
# for col in df_c.columns:
#     if col.startswith('nuclei'):
#         new_col = '_'.join([col.split('_')[0], col.split('_')[-1]])
#     else:
#         new_col = col
#     new_cols.append(new_col)
                           
# bmcc.columns = new_cols
# bmcc.head()
                 
df_c.columns = [col[:len('nuclei-xxx')] if col.startswith('nuclei') else col for col in df_c.columns]
df_mc.columns = [col[:len('nuclei-xxx')] if col.startswith('nuclei') else col for col in df_mc.columns]
df_c.columns = ['_'.join(col.split('-')) for col in df_c.columns]
df_mc.columns = ['_'.join(col.split('-')) for col in df_mc.columns]
df_c.columns = [col+'_c' if col.startswith('nuclei') else col for col in df_c.columns]
df_mc.columns = [col+'_mc' if col.startswith('nuclei') else col for col in df_mc.columns]
```

```python
df_c.columns
df_cluster_mc_c = pd.DataFrame() 
for label, df_sub in df_meta.groupby('Neuron type'):
    samples = df_sub.index.values
    df_cluster_mc_c[label+'_mc'] = df_mc[samples+'_mc'].sum(axis=1)
    df_cluster_mc_c[label+'_c'] = df_c[samples+'_c'].sum(axis=1)
    
print(df_cluster_mc_c.shape)
print(df_cluster_mc_c.head())
```

```python
print(df_cluster_mc_c.shape)
print(df_cluster_mc_c.tail())
```

```python
from CEMBA_init_singleton_ensemble import merge_bins
df_out = merge_bins(df_cluster_mc_c.reset_index(), bin_size=100000, double_xsize=True, output_file=None)
print(df_out.shape)
```

```python
df_out = df_out.set_index(['chr', 'bin'])
df_out.tail()
```

```python
df = df_out
cov = (df.filter(regex='_c$').apply(np.log10)>3)*1
gene_cov = cov.apply(np.all, axis=1)
df_filtered = df.loc[gene_cov, :]
print(df_filtered.shape)
```

```python
df_mc = df_filtered.filter(regex='_mc$')
df_c = df_filtered.filter(regex='_c$')
df_mc.columns = [col[:-len('_mc')] for col in df_mc.columns]
df_c.columns = [col[:-len('_c')] for col in df_c.columns]
df_mcc = df_mc/df_c
df_mcc.columns = [col+'_mcc' for col in df_mcc.columns]
print(df_mcc.shape)
df_mcc.tail()
```

```python
output_f = os.path.join(PATH_REFERENCES, 'Mouse_published/binc_mCH_100000_clusterwise_mcc_mouse_published.tsv')
df_mcc.to_csv(output_f, na_rep='NA', header=True, index=True, sep='\t')
```
