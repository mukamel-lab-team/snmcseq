```python
ens = 'Ens7'
context = 'CH'
ens_genelevel_path = os.path.join(PATH_ENSEMBLES, ens, 'gene_level')
engine = connect_sql('CEMBA')


# get cells and genebody paths
cells = []
genebody_paths = []
sql = '''SELECT cell_name, dataset FROM cells 
JOIN {} on cells.cell_id = {}.cell_id'''.format(ens, ens)
df_info = pd.read_sql(sql, engine, index_col='cell_name')
for dataset, df_sub in df_info.groupby('dataset'):
    cells_tmp = sorted(df_sub.index.tolist())
    paths_tmp = [os.path.join(PATH_DATASETS, dataset, 'gene_level', 'genebody_{}.tsv.bgz'.format(cell)) 
                 for cell in cells_tmp]
    cells += cells_tmp
    genebody_paths += paths_tmp
    
```

```python
cell_mcc = pull_genebody_info(ens, ens_genelevel_path, cells, genebody_paths, 
                contexts=CONTEXTS, to_file=True)

# read directly from files
```
