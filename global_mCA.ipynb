```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import os
import logging
import glob

from __init__ import *
import snmcseq_utils 
from snmcseq_utils import create_logger
```

```python
dataset = 'CEMBA_3C_171206'
dataset_path = os.path.join(PATH_DATASETS, dataset)
allc_paths = sorted(glob.glob(os.path.join(dataset_path, 'allc/allc_*.bgz')))
binc_paths = sorted(glob.glob(os.path.join(dataset_path, 'binc/binc_*.bgz')))
meta_path = os.path.join(dataset_path, 'mapping_summary_{}.tsv'.format(dataset))
```

```python
df_meta = pd.read_table(meta_path, index_col='Sample')
```

```python
contexts = CONTEXTS
res_all = []
ti = time.time()

for binc_file in binc_paths[:100]:
    print('.', end='')
    df = snmcseq_utils.read_binc(binc_file, compression='gzip')
    # filter chromosomes
    df = df[df.index.get_level_values(level=0).isin(snmcseq_utils.get_mouse_chromosomes())]

    res = {}
    res['sample'] = os.path.basename(binc_file)[len('binc_'):-len('_10000.tsv.bgz')]
    sums = df.sum()
    res['mCA'] = sums['mCA']/sums['CA']
    res['mCH'] = sums['mCH']/sums['CH']
    res['mCG'] = sums['mCG']/sums['CG']
    res_all.append(res) 
    
df_res = pd.DataFrame(res_all)
df_res = df_res.set_index('sample')
tf = time.time()
print(tf-ti)
```

```python
df_res = pd.DataFrame(res_all)
df_res = df_res.set_index('sample')
df_res = pd.merge(df_res, df_meta, left_index=True, right_index=True)
df_res.head()
```
